### Um sistema discreto estável, causal, tem como função de transferência $H(z) = cos(z^{-1})$. Elabore uma simulação computacional que calcule e trace a resposta em amplitude deste sistema para uma faixa apropriada de frequências digitais $Ω$. Comente sobre o sistema. Determine também a resposta ao impulso para $n$ de $0$ a $10$. Determine a equação de diferença que descreve o filtro FIR que aproxima o sistema $H(z)$. Para verificar o comportamento adequado, trace a resposta em magnitude do filtro FIR e compare com a resposta em magnitude do sistema original. 

#### Simulação computacional da resposta em amplitude, em Python:

```python
import numpy as np
from scipy import signal
import pylab as pl
import matplotlib.pyplot as plt


w = np.linspace(-np.pi,np.pi,500)
h = np.cos(np.exp(-1j*w))

fig = plt.figure()
plt.title('\n Resposta Frequencial de $H(z) = cos(z^{-1})$\n')
plt.plot(w, 20 * np.log10(abs(h)), 'c')
plt.ylabel('Amplitude [dB]')
plt.xlabel('Frequencia [rad]')
plt.grid()
plt.axis('tight')

plt.show()
```

#### Resposta ao impulso para $n$ de $0$ a $10$:

```python
from scipy import signal
import numpy as np
from numpy import arange
import pylab as pl
import math as m
import matplotlib.pyplot as plt

order = 5
n=arange(0,10)
b=[]

#Implementacao da serie de Taylor
for i in range(0, len(n)):
    b.append(0.0)

for i in range(0,order,2):
    if (i/2)%2 == 0:
        b[i] = (1.0 ** (i/2.0)) / m.factorial(i)
    else:
        b[i] = (-1.0 ** (i / 2.0)) / m.factorial(i)

print b

fig = plt.figure()
plt.title('\n Resposta ao Impulso de $H(z) = cos(z^{-1})$\n')
plt.stem(n, b, 'c')
plt.ylabel('h [n]')
plt.xlabel('n')
plt.grid()
plt.axis('tight')

plt.show()
```

#### Resposta em magnitude do filtro FIR:

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.linspace(-np.pi,np.pi,500)
z = np.exp(1j*w)
h = 1 - z**(-2)/2 + z**(-4)/24

fig = plt.figure()
plt.title('\n Resposta Frequencial do filtro FIR\n')
plt.plot(w, 20 * np.log10(abs(h)), 'c')
plt.ylabel('Magnitude [dB]')
plt.xlabel('Frequencia [rad]')
plt.grid()
plt.axis('tight')

plt.show()
```
